```python
import os
import json
import numpy as np
import pandas as pd
```

```python
# Directory setup
output = "./interactive/mechanics/"

# Image setup
template = 'plotly'

# Check if directory exists, if not, create it
if not os.path.exists(output):
    os.makedirs(output)   
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Function to generate position and velocity data

def constant_force(x0,v0, a, t0, tf, dt):
    
    time = []
    velocity = []
    position = []
    
    t = t0
    
    while t <= tf:
        
        time.append(t)
        
        v = v0 + a * (t - t0)
        x = x0 + v0 * (t - t0) + 0.5 * a * (t - t0)**2
        
        position.append(x)
        velocity.append(v)
        
        t += dt

    return position, velocity, time
```

# Constant Velocity Figures

```python
# Constant Velocity Data

position, velocity, time = constant_force(x0=12, v0=12, a=0, t0=0, tf=10, dt=0.1)

# Create a position vs. time plot

file1 = "CV_position_vs_time.html"
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=time, y=position, mode='lines+markers', name='Position'))
fig1.update_layout(
    template=template,
    #title='Position vs. Time',
    xaxis_title='Time (s)',
    yaxis_title='Position (m)',
    xaxis=dict(range=[0, 10]),
    yaxis=dict(range=[-10, 150])
)
fig1.add_hline(y=0, line_dash="dash", line_color="black", line_width=1)

fig1.show()

fig1.write_html(os.path.join(output, file1))

# Create a velocity vs. time plot

file2 = "CV_velocity_vs_time.html"
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=time, y=velocity, mode='lines+markers', name='Velocity'))
fig2.update_layout(
    template=template,
    #title='Velocity vs. Time',
    xaxis_title='Time (s)',
    yaxis_title='Velocity (m/s)',
    xaxis=dict(range=[0, 10]),
    yaxis=dict(range=[-1, 15])
)
fig2.add_hline(y=0, line_dash="dash", line_color="black", line_width=1)

fig2.show()

fig2.write_html(os.path.join(output, file2))
```

# Force Graphs

```python
# Define forces for each second (positive and negative)
forces = [5, 8, -5, 6, -7]  # Ns for each second
time_intervals = [0.5, 1.5, 2.5, 3.5, 4.5]

# Assign colors based on positive (light blue) or negative (light red)
bar_colors = ['#636EFA' if force > 0 else '#EF553B' for force in forces]

file7 = "net_force_vs_time_discrete.html"
fig7 = go.Figure()
fig7.add_trace(go.Bar(
    x=time_intervals,
    y=forces,
    marker_color=bar_colors,
    # text=forces,
    textposition='outside'
))

fig7.update_layout(
    template=template,
    #title='Impulse Over Time',
    xaxis_title='Time (s)',
    yaxis_title='Net Force (N)',
    yaxis=dict(range=[min(forces) - 2, max(forces) + 2]),
    bargap=0
)
fig7.add_hline(y=0, line_dash="dash", line_color="black", line_width=1)


fig7.show()
fig7.write_html(os.path.join(output, file7))
```

```python
# Create a smooth force function using sine waves
time_fine = np.linspace(0, 5, 500)
force_function = 5 * np.sin(2 * time_fine) + 2 * np.cos(3 * time_fine)

# Create Riemann sum bars - smaller width
num_bars = 40
bar_width = 5 / num_bars
time_intervals = [i * bar_width + bar_width/2 for i in range(num_bars)]
forces = [5 * np.sin(2 * t) + 2 * np.cos(3 * t) for t in time_intervals]

# Assign colors based on positive (light blue) or negative (light red)
bar_colors = ['#636EFA' if force > 0 else '#EF553B' for force in forces]

file8 = 'net_force_vs_time_smooth.html' 
fig8 = go.Figure()

# Add smooth curve
fig8.add_trace(go.Scatter(
    x=time_fine,
    y=force_function,
    mode='lines',
    line=dict(color='darkblue', width=2)
))

# Add Riemann sum bars
fig8.add_trace(go.Bar(
    x=time_intervals,
    y=forces,
    width=bar_width,
    marker_color=bar_colors
))

fig8.update_layout(
    template=template,
    xaxis_title='Time (s)',
    yaxis_title='Force (N)',
    yaxis=dict(range=[-8, 8]),
    showlegend=False
)
fig8.add_hline(y=0, line_dash="dash", line_color="black", line_width=1)

fig8.show()
fig8.write_html(os.path.join(output, file8))
```

```python
# Sinusoidal position vs time with coarse vs fine Riemann sums

# Parameters
t0, tf = 0, 10
A = 5.0  # amplitude
t_fine_pos = np.linspace(t0, tf, 1000)
position_func = A * np.sin(2 * np.pi * (t_fine_pos - t0) / (tf - t0))

# Coarse (big step) Riemann sum
N_big = 25
width_big = (tf - t0) / N_big
t_big = [t0 + (i + 0.5) * width_big for i in range(N_big)]
p_big = A * np.sin(2 * np.pi * (np.array(t_big) - t0) / (tf - t0))

# Fine (smaller step) Riemann sum
N_small = 80
width_small = (tf - t0) / N_small
t_small = [t0 + (i + 0.5) * width_small for i in range(N_small)]
p_small = A * np.sin(2 * np.pi * (np.array(t_small) - t0) / (tf - t0))

# Colors
coarse_color = '#636EFA'   
fine_color   = '#EF553B'  

file9 = 'sinusoidal_position_vs_time_coarse_fine.html'
fig9 = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.12,
    subplot_titles=(
        f'Coarse (Big t) (N={N_big})',
        f'Fine (Small t) (N={N_small})'
    )
)

# Coarse (top)
fig9.add_trace(go.Scatter(x=t_fine_pos, y=position_func, mode='lines',
                          line=dict(color='darkblue', width=2),
                          name='Position'), row=1, col=1)
fig9.add_trace(go.Bar(x=t_big, y=p_big, width=width_big,
                      marker_color=coarse_color,
                      name='Coarse Bars'), row=1, col=1)
fig9.add_hline(y=0, line_dash="dash", line_color="black", line_width=1, row=1, col=1)

# Fine (bottom)
fig9.add_trace(go.Scatter(x=t_fine_pos, y=position_func, mode='lines',
                          line=dict(color='darkblue', width=2),
                          name='Position', showlegend=False), row=2, col=1)
fig9.add_trace(go.Bar(x=t_small, y=p_small, width=width_small,
                      marker_color=fine_color,
                      name='Fine Bars', showlegend=False), row=2, col=1)
fig9.add_hline(y=0, line_dash="dash", line_color="black", line_width=1, row=2, col=1)

fig9.update_layout(
    template=template,
    #title='Sinusoidal Position vs Time: Coarse vs Fine Riemann Sums',
    barmode='overlay',
    xaxis2_title='Time (s)',
    yaxis_title='Position (m)',
    yaxis2_title='Position (m)',
    showlegend=False
)

fig9.show()
fig9.write_html(os.path.join(output, file9))
```

```python
# Build an index.html listing all HTML figures in the output folder

index_name = "index.html"
html_files = sorted(
    [f for f in os.listdir(output) if f.endswith(".html") and f != index_name],
    key=str.lower
)

def nice_title(filename):
    name = os.path.splitext(filename)[0]
    return name.replace("_", " ").title()

# Generate list items from files
items = "\n".join(
    f'          <li><a href="{fname}" target="_blank">{nice_title(fname)}</a>'
    f' <span class="filename">{fname}</span></li>'
    for fname in html_files
)

# Generate body content (list or empty message)
body_content = (
    "<ul class='file-list'>" + items + "</ul>"
    if html_files
    else "<div class='empty'>No HTML figures found.</div>"
)

# Load CSS template and CSS variables
with open('./static/index_styles.css', 'r') as f:
    css_template = f.read()

with open('./static/css_vars.json', 'r') as f:
    css_vars = json.load(f)

# Apply CSS variables to template
css_content = css_template % css_vars

# Load HTML template
with open('./static/index_template.html', 'r') as f:
    html_template = f.read()

# Build final HTML
content = html_template % {
    'css_content': css_content,
    'body_content': body_content
}

# Write to file
index_path = os.path.join(output, index_name)
with open(index_path, "w", encoding="utf-8") as f:
    f.write(content)

print(f"Index written to: {index_path} ({len(html_files)} links)")
```
